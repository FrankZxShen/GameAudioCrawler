```python
# 爬取角色音频
import os
import time
# from bs4 import BeautifulSoup
import requests
from tqdm import *
import re
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait

def download_audio(audio_l,c_name): 
    if os.path.exists(c_name): 
        print(c_name," 已存在")
        return
    else:
        os.mkdir(c_name) 
    
    for i in tqdm(audio_l): 
        audio = requests.get(i) 
        audio_name = i.split('/')[12] 
        with open(c_name + '\\' + audio_name, 'wb') as f: 
            f.write(audio.content)

def get_mp3_names(html):
    rex = re.compile('<audio src="(.*?)"></audio>') 
    audio_list1 = re.findall(rex, html)
    audio_list2 = ['https:'+ a[:a.find('"')] for a in audio_list1]
    audio_list3 = []
    for a in audio_list2:
        if a[-2:] == 'og':
            audio_list3.append(a+'g')
        elif a[-2:] == 'mp':
            audio_list3.append(a+'3')
        elif a[-2:] == 'wa':
            audio_list3.append(a+'v')
        else:
            audio_list3.append(a)
    print(audio_list3)
    # print(audio_list2)
    return audio_list3

print("开始爬取...")
print("#####################################################")

############################################################################################
# 开始执行爬取
os.chdir('./audio/')
for url, name in zip(c_url,c_name):
#############################################################################################
    print("角色名：",name)
    try:
        browser = webdriver.Chrome()
        browser.minimize_window()
        browser.get(url)
        wait = WebDriverWait(browser,10) #等待变量
        html = browser.page_source
        audio_list = get_mp3_names(html)
        download_audio(audio_list, name)
        browser.quit()
    except Exception as e:
        print("爬取 "+ name + " 出现错误：",e)
    time.sleep(2)

print("#####################################################")
print("爬取结束...")
```
